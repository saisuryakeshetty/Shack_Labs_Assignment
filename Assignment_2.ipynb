# Import Libraries

```python
import numpy as np
import pandas as pd
```

# Read CSV files

```python
amz_df = pd.read_csv('DS - Assignment Part 2 data set/amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
flipkart_df = pd.read_csv('DS - Assignment Part 2 data set/flipkart_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
amz_df.head()
```

```python
flipkart_df.head()
```

# Merge DataFrames

```python
df = pd.merge(flipkart_df,amz_df,how = 'inner',on='uniq_id')
```

```python
df.head()
```

# Label Indexing

```python
df_match = df.loc[:,['product_name_x','retail_price_x','discounted_price_x','product_name_y','retail_price_y','discounted_price_y']]
```

```python
df_match.head()
```

# Rename Columns

```python
dic = {'product_name_x':'Product name in Flipkart','retail_price_x':'Retail Price in Flipkart','discounted_price_x':'Discounted Price in Flipkart',
       'product_name_y':'Product name in Amazon','retail_price_y':'Retail Price in Amazon','discounted_price_y':'Discounted Price in Amazon'}
df_match.rename(columns = dic,inplace =True)
```
